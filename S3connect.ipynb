```python
from aws_access import access_key, secret_access_key
import boto3
import os
```

```python
client = boto3.client('s3',aws_access_key_id = access_key,
                      aws_secret_access_key = secret_access_key)
```

```python
# if there is a CSV in this working directory then upload to s3
# if file name exists in s3, it will be overwritten
for file in os.listdir():
    if '.csv' in file:
        bucket = 'automatic-csv-uploader'
        bucket_dir = 'python/'+ str(file)
        client.upload_file(file, bucket, bucket_dir)
        
```
